```python
%pylab inline
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import numpy as np

dt = [('Date (MDJ)', '<i8'),('Data','<f8')]

google_data = np.loadtxt('google_data.txt',dtype=dt,skiprows=1)
yahoo_data = np.loadtxt('yahoo_data.txt',dtype=dt,skiprows=1)
ny_temps = np.loadtxt('ny_temps.txt',dtype=dt,skiprows=1)

# Open a new figure and add the title
f = plt.figure()
f.suptitle('New York Temperature, Google, and Yahoo!', fontsize=16)

# Plot the stock data against the left axis
l1, = plt.plot(google_data['Date (MDJ)'],google_data['Data'])
l2, = plt.plot(yahoo_data['Date (MDJ)'],yahoo_data['Data'],'m')

# Assign to a variable the left axis handle
axL = plt.gca()

# Label the left y-axis and the x-axis
axL.set_ylabel('Value (Dollars)')
axL.set_xlabel('Date (MDJ)')

# Assign y-limits to the left axis
axL.set_ylim(-20,770)

# "Twin" the x-axis so that multiple y-axes can be used
axR = axL.twinx()

# Plot the N.Y. temperature data against the right y-axis
l3, = plt.plot(ny_temps['Date (MDJ)'],ny_temps['Data'],'r--')
axR.set_ylim([-150,100])
axR.set_ylabel('Temperature ($\degree$F)')
l3.set_dashes([2,2])

# Make the legend from the two axes' data
handles = [l2,l1,l3]
labels = ['Yahoo! Stock Value','Google Stock Value','N.Y. Mon. High Temp']
plt.legend(handles,labels,loc=6,frameon=False)

# Set x limits
axL.set_xlim(48800,55610)

# Show minor tick marks
axL.xaxis.set_minor_locator(MultipleLocator(200))
axL.yaxis.set_minor_locator(MultipleLocator(20))
axR.yaxis.set_minor_locator(MultipleLocator(10))
```
