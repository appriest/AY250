```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# Set up the types of the loaded text files
dt = [('frac_followed','<f16'),('frac_observed','<f16'),('frac_observed_uncertainty','<f16')]

# Load the text files
eff = np.loadtxt('HW31_example/Efficiency.txt',skiprows=1,dtype=dt)
pur = np.loadtxt('HW31_example/Purity.txt',skiprows=2,dtype=dt)

# Plot the two graphs on subplots
f,(ax1,ax2) = plt.subplots(1,2,figsize=(12,6),dpi=100)
ax1.plot(eff['frac_followed'],eff['frac_observed'],'black')
ax1.plot([0,1],'black')
ax2.plot(pur['frac_followed'],pur['frac_observed'],'black')
ax2.plot([pur['frac_observed'][-1],pur['frac_observed'][-1]],'black')
f.show()

# Set the limits
ax1.set_xlim(-.05,1.05)
ax1.set_ylim(-.05,1.05)
ax2.set_xlim(-.05,1.05)
ax2.set_ylim(-.05,1.05)

# Labels
fd_axis={'family' : 'normal','weight' : 'normal','size'   : 12}
fd_title={'family' : 'normal','weight' : 'bold','size'   : 16}
ax1.set_ylabel('Fraction of High (z>4) GRB\'s Observed', fontdict=fd_axis)
ax1.set_xlabel('Fraction of GRB\'s Followed Up', fontdict=fd_axis)
ax1.set_title('Efficiency',fontdict=fd_title)
ax2.set_ylabel('Percent of Observed GRB\'s That Are High z (z>4)', fontdict=fd_axis)
ax2.set_xlabel('Fraction of GRB\'s Followed Up', fontdict=fd_axis)
ax2.set_title('Purity',fontdict=fd_title)

# Set ticklines outside and turn right and top ticks off
for tick in ax1.get_xticklines():
    tick.set_marker(mpl.lines.TICKDOWN)
    
for tick in ax1.get_yticklines():
    tick.set_marker(mpl.lines.TICKLEFT)
    
for tick in ax2.get_xticklines():
    tick.set_marker(mpl.lines.TICKDOWN)
    
for tick in ax2.get_yticklines():
    tick.set_marker(mpl.lines.TICKLEFT)
    
ax1.xaxis.tick_bottom()
ax1.yaxis.tick_left()
ax2.xaxis.tick_bottom()
ax2.yaxis.tick_left()

# Annotations
ax1.annotate('Follow up 20% of\nburts to capture ~55%\nof high-z events', xy=(0.2, 0.55), xytext=(0.43, 0.25),
            arrowprops=dict(facecolor='black', width=0.9, shrink=0.05, headwidth=7),
            horizontalalignment='left',verticalalignment='center')

ax1.annotate('Random\nGuessing', xy=(0.8, 0.8), xytext=(0.8, 0.6),
            arrowprops=dict(facecolor='black', width=0.9, shrink=0.05, headwidth=7),
            horizontalalignment='center',verticalalignment='center')

ax2.annotate('If 20% of events are\nfollowed up, ~40% of\nthem will be high z', xy=(0.2, 0.38), xytext=(0.2, 0.8),
            arrowprops=dict(facecolor='black', width=0.9, headwidth=7),
            horizontalalignment='center',verticalalignment='center')

ax2.annotate('Random\nGuessing', xy=(0.6, pur['frac_observed'][-1]), xytext=(0.8, 0.6),
            arrowprops=dict(facecolor='black', width=0.9, headwidth=7),
            horizontalalignment='center',verticalalignment='center')

# Vertical marker lines
ax1.axvline(x=0.2,ymin=0,ymax=0.55,color='black',linestyle=':',dash_capstyle='round',linewidth=2)
ax2.axvline(x=0.2,ymin=0,ymax=0.39,color='black',linestyle=':',dash_capstyle='round',linewidth=2)

# Patch for error regions
eff_polyLims = zip(eff['frac_followed'],eff['frac_observed']+eff['frac_observed_uncertainty'])+zip(eff['frac_followed'],eff['frac_observed']-eff['frac_observed_uncertainty'])[::-1]
eff_poly = mpl.patches.Polygon(eff_polyLims,facecolor='0.9',edgecolor='0.5')
ax1.add_patch(eff_poly)

pur_polyLims = zip(pur['frac_followed'],pur['frac_observed']+pur['frac_observed_uncertainty'])+zip(pur['frac_followed'],pur['frac_observed']-pur['frac_observed_uncertainty'])[::-1]
pur_poly = mpl.patches.Polygon(pur_polyLims,facecolor='0.9',edgecolor='0.5')
ax2.add_patch(pur_poly)
```
